# train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = dataset.drop('prognosis', axis=1)
y = dataset['prognosis']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# ecoding prognonsis
le = LabelEncoder()
le.fit(y)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
y
```

```python
X_train, X_test, y_train, y_test.shape
```

# Training top models

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Model 1 : Support Vector Classifire Algorithm

```python

# Create a dictionary to store models
models = {
    'SVC': SVC(kernel='poly', C=0.5, degree=5),
}

# Loop for the models train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Test the model
    predictions = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy}")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))

    print("\n" + "="*40 + "\n")
    
plt.figure(figsize=(5, 3))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for SVC")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

Model 2 : Random Forest Algorithm

```python

# Create a dictionary to store models
models = {
    'RandomForest': RandomForestClassifier(
        n_estimators=80,
        max_depth=2,
        min_samples_split=4,
        min_samples_leaf=2,
        bootstrap=True,
        oob_score=True,
        random_state=42),
}

# Loop for the models train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Test the model
    predictions = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy}")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))

    print("\n" + "="*40 + "\n")
```

Model 3:Gradient Boosting Classifier

```python
# Create a dictionary to store models
models = {
    'GradientBoosting': GradientBoostingClassifier(
        n_estimators=80, 
        learning_rate=0.01,
        max_depth=7,
        subsample=0.2),
}

# Loop for the models train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Test the model
    predictions = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy}")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))

    print("\n" + "="*40 + "\n")
```

Model 4 : DecisionTree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
# Create a dictionary to store models
models = {
    'dt_classifier': DecisionTreeClassifier(
        random_state=42, 
        criterion="gini",
        max_depth=30,
        min_samples_split=50,
    )
}

# Loop for the models train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Test the model
    predictions = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy}")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))

    print("\n" + "="*40 + "\n")
```
